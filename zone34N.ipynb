```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib import colors
import folium
from shapely.geometry import Point
from shapely.geometry.polygon import Polygon
from pyproj import Proj, transform
import numpy as np
```

```python
gps = Proj(init='epsg:4326')
zone34N = Proj(init='epsg:23034')
```

```python
a = Point(-460650.21 + 1000000, 6421187.79+100000)
```

```python
list = []

grid_size = 200_000 # 200 km
i = 0
start_x = -460650.21
end_x = 1125674.78

start_y = 6421187.79
end_y = 7979913.77
j = 0
while start_y + j*grid_size < end_y:
    while start_x + i*grid_size < end_x:
        list.append((transform(zone34N, gps, start_x + i*grid_size, start_y+j*grid_size), transform(zone34N, gps, start_x + (i-1)*grid_size, start_y+(j-1)*grid_size)))
        i += 1
    i = 0
    j += 1
```

```python

#Create color map
color_map = [colors.rgb2hex(cmap[np.random.randint(0, len(list))]) for i in range(len(list))]

for i in range(len(list)):
    color = colors.rgb2hex(color_map[i])
    folium.Rectangle(bounds=((list[i][0][1], list[i][0][0]), (list[i][1][1], list[i][1][0])), radius=1, color=color,
                        fill_color=color, ).add_to(m)
```

```python
df_locations = pd.read_json("./data/locations_new.json")
df_locations.describe
df_locations['box_index'] = [None]*df_locations['lat'].size
```

```python
for index in df_locations.index:
    point = Point(transform(gps,zone34N, df_locations['lng'][index], df_locations['lat'][index]))
    #Check if point is inside any of the boxes in list 
    for i in range(len(list)):
        if point.within(Polygon([list[i][0], list[i][1], (list[i][1][0], list[i][0][1]), (list[i][0][0], list[i][1][1])])):
            df_locations['box_index'][index] = i
            break
```

```python
#add all the points to the map with the correct box color
for index in df_locations.index:
    if(df_locations['box_index'][index] == None):
        continue
    color = colors.rgb2hex(color_map[df_locations['box_index'][index]])
    folium.CircleMarker(location=(df_locations['lat'][index], df_locations['lng'][index]), radius=1, color=color,
                        fill_color=color, ).add_to(m)
```
