```python
import pandas as pd
import matplotlib.pyplot as plt
import folium
```

```python
df = pd.read_json('./data/locations.json')
```

```python
m = folium.Map((64.9701, 13.509217), zoom_start=4)
[folium.CircleMarker(location=(row[1]['lat'], row[1]['lng']), radius=1, color='#0080bb',
                     fill_color='#0080bb', ).add_to(m) for row in df.iterrows()]
m
```

```python
df_plt = df.sample(frac=1)
fig = plt.figure(figsize=(8, 8))
fig.set_facecolor('white')
plt.scatter(df_plt['lng'], df_plt['lat'], s=1)
plt.scatter(10, 60)
plt.grid()
```
