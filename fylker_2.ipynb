```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib import colors
import folium
from shapely.geometry import Point
from shapely.geometry.polygon import Polygon
from pyproj import Proj, transform
import os
```

```python
df = pd.read_csv('./data/locations_county.csv').iloc[:3600]
lst = os.listdir("./data/images")
df = df[df['filename'].str.contains('|'.join(lst)) == True]
targets = df["fylkesnummer"].to_numpy()
```

```python
df = df.reset_index()
```

```python
viridis = cm.get_cmap('viridis', 12)
cmap = viridis(range(11))

m = folium.Map((64.9701, 13.509217), zoom_start=4)
for i in df.index:
    if i < 2100: 
        continue
    if df['fylkesnummer'][i] == None:
        continue
    color = colors.rgb2hex(cmap[int(df['fylkesnummer'][i])])
    folium.CircleMarker(location=(df['lat'][i], df['lng'][i]), radius=1, color=color,
                     fill_color=color, ).add_to(m)
```
